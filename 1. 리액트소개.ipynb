# 1차시 리액트 소개

### 1) SPA와 컴포넌트 모델 소개(Introduction SPA & Components)

- React : The libary for web and native user interfaces
    - 화면을 만드는 기술, 프레임워크가 아닌 라이브러리

- 전형적인 웹앱 => 멀티페이지
    - 옛날 방식의 멀티 페이지


- SPA - 싱글 페이지 앱
    - 화면의 일부분만 갱신되도록 구현, 화면의 일부분 = 컴포넌트
    - 하나의 화면을 갱신하기위해서 서버에서는 데이터만 보내줌



- SPA 어플리케이션 개발 시 특징
    - 데이터와 UI 동기화 
        - (데이터가 바뀌면 해당 UI가 변경)
    - DOM 조작 - DOM은 느리다 
        - (리소스를 많이 잡아 먹음) 
        - => SPA에서는 가상 DOM을 사용하기도 한다 기존의 돔에서 사본의 DOM을 바꾸고 바뀐부분만 원본에 반영, (메모리는 많이 차지, 무조건 빠른건 아니다)
    - HTML에 데이터 바인딩하기
        -Interpolation
        

- 모던 웹 아키텍처 지원 프론트엔드 프레임워크의 특징
    
    - 컴포넌트(화면)는 데이터와 레이아웃으로 구성되어 있다.
        - 데이터를 state라고 한다.
        - state는 화면의 엘레먼트(태그)에 바인딩 되어 있다.
        - state 값이 바뀌면 바인딩된 태그(화면)도 자동으로 다시 렌더링 된다.

- 컴포넌트 기반 프론트엔드 개발
    - 작고 단단한 컴포넌트를 만드는 것
    - 이렇게 만들어진 컴포넌트간의 관계를 정의하고 유기적으로 연결하는것

컴포넌트가 어떻게 재사용이 가능하고, 서로 유기적으로 연결되어있게 설계하는지가 핵심

## 2) REACT 특징

- 컴포넌트를 이용해 화면을 만드는 기술


- JSX(HTML과 유사)와 코드(JS, TS)를 이용해 컴포넌트를 작성<br>


- 변경에 즉각 대응하여 화면이 변한다.<br>
<br>
※ 선언적 프로그래밍: 개발자는 데이터만 신경쓰면 된다. 
<br>    (데이터만 바꾸면 React로 화면이 자동으로 바뀐다)<br>
※ 리액트 : 라이브러리 이기때문에 부가적인 기능을 제공하지 않는다.<br>
※ 이 방안을 개선하기 위해, 나온 것이 리액트 네이티브<br>
※ 리액트 네이티브 : 화면 만드는 기술과 실무에서 프로덕션의 앱을 만들때 부가적인 기술들을 모아서 풀스텍이 가능하게 서버사이드 렌더링 등의 서버 기능까지 사용할 수 있도록 만들어진 메타프레임워크 (유사 Next.js)

- 프레임워크를 이용해 풀스텍 개발이 가능


- 다양한 플랫폼 지원


리액트 네이티브 웹, 앱, 데스크탑 어플리케이션도 만들 수 있도록 지원함.

### 3)프론트엔드 프레임워크 비교
- React, Vue, Angular, Svelte

- 라이브러리 VS 프레임워크
    - 라이브러리 (React)
        - 개발자가 사용
        - 자유도 높음, 규격화 낮아짐

    - 프레임워크 (Angular, Vue)
        - 프레임워크에 맞춰 개발
        - 자유도 낮음, 규격화 높아짐(가독성 높아짐, Readability)

    - 라이브러리 VS 프레임워크
        - 애플리케이션 복잡도가 증가되면 프레임워크가 적합

- Angular VS React
    - Angular (typescript 기본개발)
        - From Google
        - MVVM 프레임워크
        - AOT 렌더링, 라우터, CLI
        - 기본적으로 TypeScript 채택
        - 디펜던시 인젝션 기능
        - 글로벌 상태관리 시스템 부재 → Flux, NgRX 채택 필요
    
    - React, Vue (ES6 기본 개발)
        - 컴포넌트 기반 개발에 중점을 둠 <br>→ 프레임워크 도입에 노력과 시간을 줄여줌


- React VS Vue
    - 2013년 Facebook 에서 오픈 릴리즈 됨
        - JSX 패턴(Javascript에서 HTML과 CSS를 작성하는 방법)
        - 모듈 방식의 애플리케이션 구축

    - Vue → Svelte
        - 모듈식 코딩의 핵심을 취함
        - JavaScript 단일 파일에서 HTML, CSS 요소를 포함(SFC)
        - 모듈식 개발 방식의 단순화 <br><br>

        ※ 가상 DOM을 사용하지 않고 성능이 더 좋아짐<br>
        ※ 가장 대중적이고 레퍼런스가 많은것(안정적이다)을 선호하면 React<br>
        ※ 직관적이고 사용하기 쉬운것을 선호하면 Vue<br>
        ※ Svelte는 대중성이나 레퍼런스가 낮지만 기술적으로는 한단계 더 진화된 단계이다.
