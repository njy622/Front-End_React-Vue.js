# 7차시. State

### 1) State, 상태
- 컴포넌트는 화면이고, 화면은 데이터를 사용한다.
- 데이터는 원본이 있고, 사본이 있다.
    - 원본: State, 사본: Props


- 원본은 변경 가능(mutable), 사본은 변경 불가능(immutable)
    - 원본이 변경되면 사본이 같이 변한다.

### 데이터가 변하면 해당 데이터를 사용하는 화면이 즉각 변한다.

State는 따로 관리하고 데이터는 props를 통해 데이터를 전송하는 과정으로 개발한다

### 2) 참고: 선언적 프로그래밍?
- 명령형 프로그래밍 방식
    - imperative Programming
    - 데이터가 변경되면, 변경된 데이터로 화면(DOM)을 직접 변경한다.
    - programmimatic

- 선언적 프로그래밍 방식
    - Declarative Programming
    - 데이터가 변경되면, 변경된 데이터를 사용하는 화면이 자동적으로 변경된다.
    - reactive

### 3) State가 변하면 화면이 다시 그려지는 원리
- 렌더링 트리거
    - 최초 화면이 로딩 될 때
    - State 값이 변경되었을 때

- 렌더링 수행
    - 변경될 태그들을 계산 → 실제 화면에 반영되지 않음 (메모리 상에서 계산되고 저장된 상태)
    - 외부의 접근 없이 내부의 데이터만을 이용해서 수행되어야 함

- DOM 커밋 
    - 실제 화면에 반영되어 화면에 그려짐

### 4) useState Hook
- state 값 관리를 위한 useState()
    - 함수형 컴포넌트로 state 관리 가능 (react 16.8에서 Hook 도입)

- const [number, setNumber] = useState(0);
    - number : state 이름
    - setNumber : number state 값 변경 함수
    - 0 : number state의 기본값

※ 원래 리액트 컴포넌트를 만들때,
16.8 이전까지는 컴포넌트 만드는 방식이 두가지였음
1) 클래스 컴포넌트 = state를 쓴다는것
2) 함수 컴포넌트

이원화 되어 있는 형태를 좀 더 효율적인 함수형 컴포넌트로 일원화하고자,
함수형 컴포넌트에도 state를 사용할 수 있도록 만든 것이 HOOK

### 5) 리액트에서의 상태(State) 관리
- 리액트(hook)을 통해 상태를 관리
