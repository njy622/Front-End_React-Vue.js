# 4차시. Props

### 1) 자식 컴포넌트에게 Props 전달
- 컴포넌트 간 속성 전달은 반드시 부모에서 자식 컴포넌트로만 가능<br>


### 2) React - props
- props - 자식 컴포넌트에 데이터를 넘겨주자!


- 전달된 props 사용하기
1) 전달된 이름으로 사용하는 방법
2) 구조분해 형태로 풀어서 사용하는 방법(이 방법을 많이사용)


- props 기본값 사용하기


- 스프레드 연산자 사용하여 props 전달
    - props의 형태도 데이터가 넘어갈때 Object로 넘김 같은이름으로 넘길 수 있음으로 
    - 스프레드 연산자를 이용해서 간결한 문법으로 바꿔줄 수 있음
    

- Children을 이용하여 컨텐츠를 전달
    - 보통 props는 데이터를 전달
    - 데이터 대신 콘텐츠(태그의 집단)를 넘길 수도 있음
    - 콘텐츠를 받을때는 Children을 이용해서 받음




- 해당 계층 구조로 설계할 예정(실습)




부모의 데이터를 자식에게 가져와서 쓸 수 있음




- 부모의 props를 직접 변수명을 파라메터로 넣어서 사용할 수도 있음


- props가 여러개일때, 명칭과 값을 동일하게 넣는경우, {...props}를 이용해서 한번에 가져올 수 있음



- 카드라는 컴포넌트 템플릿역할을 하는 것을 중간에 넣어 설계한것을 실습<br>
- Children 실습<br>
