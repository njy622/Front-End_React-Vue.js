# 13차시 컴포넌트 라이프사이클 이론실습 서버에서 데이터를 갖고 오기 이론, fake api todos 실습

## Life Cycle (of component) Hooks
### 1) 라이프 사이클 훅 등록
- 컴포넌트가 초기 렌더링을 완료하고 호출되는 함수
    - onMounted
    - cf) onBeforeMount - 컴포넌트가 마운트 되기 전 호출

- 화면이 업데이트 된 후 호출되는 함수
    - onUpdated
    - cf) onBeforeUpdate - 화면(컴포넌트) 업데이트 직전 호출
    

※ 자동으로 관리되어지는 컴포넌트(누가? 리액트, 뷰)<br>
※ 라이프사이클을 메모리로 저장하고 렌더링하고 내린다.<br>
※ 콜백함수 = 훅

### 2) Vue Lifecycle Hooks - Managed
- Vue 컴포넌트의 라이프사이클 이벤트 (호출되는 시기)
    - beforeCreats: 데이터와 메소드 사용 불가 (컴포넌트가 생성되기 전)
    - created: 데이터와 메소드 사용 가능(컴포넌트가 생성된 후)
    - beforeMount:(렌더링 되기 직전)
    - mounted: (렌더링 된 후)
    - beforeUpdate:(업데이트 되기 직전)
    - updated:(업데이트 된 후)
    - beforeDestory: (디스토리하기 전)
    - destoryed: (디스토리된 후)
    

### 실습

- onMounted 함수안에 넣으면 렌더링 된 직후 바로 데이터가 뜸
