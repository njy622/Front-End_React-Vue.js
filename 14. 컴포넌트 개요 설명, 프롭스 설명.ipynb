# 14차시. 컴포넌트 개요 설명, 프롭스 설명

### 1) 컴포넌트 개요
- Vue 애플리케이션 모듈화
    - ★ 컴포넌트 계층구조(hierachical architecture)와 중첩화(nesting)
- 컴포넌트 간 데이터 전달
    - 부모 => 자식 : Props
    - 자식 => 부모 : custom events
- 슬롯과 네임드 슬롯
- Mixin
- Global VS Local 컴포넌트
- 컴포넌트 Lifecycle Hooks

### 2) 컴포넌트 트리
- 컴포넌트 = 화면 구성요소
    - 컴포넌트는 트리 구조 또는 중첩된(nested) 구조
    

※ 컴포넌트 => 컴포지션 (사용관계, 포함관계)
※ 일부 HTML태그를 포함해서 컴포넌트를 표현할 수 있다

### 3) 컴포넌트 작성
- SFC(.vue 확장자)를 사용하여 별도 파일에 Vue 컴포넌트 정의


### 실습
- defineProps를 이용하여, 부모의 변수를 자식의 컴포넌트에 받아오는 방법



- 데이터의 원본이있는 부모컴포넌트 App.vue에서 선언한 데이터를 자식 컴포넌트 태그에 바인딩해서 전달하여 자식 컴포넌트에서 defineProps로 전달받음
- defineProps는 배열의 형태로 전달받을 수 있으며, 여러개의 props를 전달 받을 수 있다.
    -   defineProps(['title', 'likes'])
- Props를 전달받을땐, 타입도 전달할 수 있음. 
    - defineProps({ title: String, likes: Number })
    - 타입체킹 방식으로 이용가능
- 이 과정에서 부모컴포넌트의 데이터가 변경되면 바인딩된 자식 컴포넌트에서의 props도 리렌더링 됨
