### 2) 첫 번째 리액트 앱

- 프로젝트 생성
    - 스캐폴딩 툴, Scaffolding
        - 기본 프로젝트 시작에 필요한 코드와 환경을 자동으로 세팅해주는 툴
        - 프로젝트를 만들어주는 툴

    - 발전
        - Angular가 2016년 발표되면서 Angular-CLI라는 스캐폴딩 툴 소개
        - react - create-react-app 이라는 스캐폴딩 툴 발표
        - Vue.js - VUE CLI를 포함 → Vue3에서 스케폴딩 툴 변경(Vite)

    - 현재 프로젝트는 모두 스캐폴딩을 통해서 프로젝트를 생성하고 개발 수행
        - 장점 동일한 코드의 구조와 아키텍처로 개발, 가독성이 높아짐

- Vite -프로젝트 스캐폴딩 <br>
※ Vue를 만든 개발자가 만든것으로 Vue와 많이 채택되어있음

### 실습 - vite를 이용한 프로젝트 만들기
설치하기 VS 확장 마켓플레이스에서


ctrl + shift + p : 명령 팔레트 <br>
terminal 쳐서 터미널 실행하기


- src : 개발자들이 주로 이용하는 폴더<br>
- public : 나중에 프로젝트를 배포할때 이용하는 폴더<bt>
- 위의 폴더 하단의 파일들 : 프로젝트 설정파일들<br>
    - package.json에서 얻을 수 있는 정보:
        - dependencies: 리액트 버전확인 가능
        - Scripts : npm run이라는 명령어가 작성되어있음
        
        
        ※ package.json파일은 이 프로젝트에 관해서 전체적인 내용이 들어가 있는 파일, npm으로 프로젝트를 만들면 root에 해당 파일이 생성됨
        ※ vite도 npm으로 프로젝트 생성하며, 다른 것도 비슷함

- 먼저 이 프로젝트에서 의존하고 있는 라이브러리 설치해야함
- cmd창에 : npm install 입력하여, node_modules 생성하기

<br>

- 설치완료 후 dev 서버 실행
    - cmd : npm run dev
    

    - 초기 안내 화면 만들어져있음
    

- src 폴더 파악하기
    - 확장자는 jsx 활용(웹표준 코드가 아닌 리액트코드임)<br>
    
    - 배포환경, 운영환경을 웹표준 코드로 바꿔주는 것(빌드)을 Vite(빌드툴)가 자동으로 해준다

    - main.jsx 파일은 제일 먼저 실행되는 파일
        

자동 저장 기능 켜놓으면 개발화면이 바로 리렌더링 되어 변경됨
