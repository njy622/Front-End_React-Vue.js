### 2) Directives, 지시문
- v-text: {{ }} 대신 사용
- v-once: 컴포넌트를 초기에 한번만 렌더링(static)
- v-html: 태그를 포함해서 렌더링 가능
- v-bind(:src): 이미지의 URL 을 바인딩 할 때
- v-if: 조건문 사용시
- v-show: 해당 태그(엘레먼트)가 보여질지 여부를 true/false 설정
- v-for: 반복문 사용시


※ v-if, v-for, v-show는 자주 이용
