# 9차시. Vue이점과 SFC, volar설치 실습, 스캐폴딩이론, vscode설정

### 1) Vue.js란?




- 3가지 특징(리액트와 대비)
    - Approachable : 쉽다
    - Performant : 성능 (리액트와 차이가 많지 않음)
    - Versatile : 다목적 기능이 많다
    - Vue에서 여러가지 기능이 제공되어있어, 자바스크립트를 깊게 알지 않아도 가능, 리액트는 자바스크립트를 깊게 들어감

### 2) 프로젝트에 Vue를 사용할 때의 이점
- 프론트엔드 개발은 배우기 쉽지만 마스터하기 어려운 패턴
    - 신규 및 베테랑 개발자 모두에게 사용 가능

- 개발자는 최적화 되고 성능이 뛰어난 프레임 워크를 사용하여 모든 크기의 동적 프론트엔드 애플리케이션을 구축가능

- 단일 파일 구성 요소(SFC) 패턴 사용
    - 개발 과정을 단순화
    - Vue가 Angular를 단순화시킴

- Pinia - 유연한 글로벌 상태 패턴을 생성

### 3) Vue 단일 페이지 패턴(SFC)
- SFC (Single File Component) 패턴
    - 작업을 더 작은 청크 단위로 분할
    - DRY(Don't Repeat Yourself)
    <br><br>
    
    <br><br>
    - Angluar는 각각의 컴포넌트마다 파일을 분리해놓여 있어, 관리는 쉬울 것 같아 보이지만 실무적으로는 복잡함
