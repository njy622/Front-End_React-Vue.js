# 10차시 스케폴딩 디렉토리 설명, 데이터바인딩 설명실습, REF

### 싱글파일 컴포넌트
- vue라는 확장자의 파일이며 하나의 컴포넌트.
- App.vue => root 컴포넌트
- App.vue 다른 컴포넌트를 사용함



1) 데이터(States)와 데이터 바인딩
- 화면은 UI와 데이터로 구성됨
    - States를 통해 Vue 템플리트 내에서 사용하는 모든 정보를 저장 가능
    - 이 데이터가 변경되면 해당 템플릿이 자동으로(반응적으로) 업데이트 됨
        - Vue는 데이터(state)를 리액티브(reactively)하게 관리
        

- 데이터는 자바스크립트에서 관리하고,
- 템플릿에서는 변수로 데이터를 바인딩함


- 데이터를 바인딩 하는 이유?
    - 프론트엔드의 아키텍쳐는 SPA(페이지가 하나)
    - App.vue => 제일 최초로 화면이 보여지며, 그다음부터는 데이터만 주고받음
    - 화면이 바뀐다는 것은 데이터가 바뀐다는 것,
    - 서버로 데이터를 요청해서 데이터가 응답이 와서 데이터로 화면을 계속 바꾸는 작업을 해야하는데, 과거에는 하나하나 프로그래미틱한 방식으로 전달했었음.
    - 바인딩되었다? 변수명의 데이터가 바뀌면 바인딩되어있는 화면이 자동으로 바뀐다.
