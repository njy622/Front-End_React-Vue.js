# 15차시. 데이터위로 전달 슬롯, 라우터(이론)

- 데이터를 아버지가 자식에게 넘겨주는 방식은 Props
- 자식의 데이터를 부모에게 넘겨주는 방식은 이벤트를 통해 전달

### 1) 이벤트를 통한 데이터 전달
- 자식 → 부모에게 데이터를 전달 시
    - 이벤트 발생(emit)과 이벤트 듣기(listening)
    - $emit : 이벤트 발생 내장 함수
    <br><br>
    
    <br><br>
    - v-on: 또는 @: 이벤트 리슨
    <br><br>
    
    <br><br>

※ 이벤트를 발생시킬 때, 데이터를 함께 전달하면서 부모에게 전달


### 실습
- year라는 props를 1씩 증가시키려면, 부모의 props에서 데이터를 변경시켜야 함으로, 이벤트를 자식 컴포넌트에서는 발생시켜야함


- $event => $emit()으로 이벤트 발생시킴
- $emit(발생시킬 이벤트이름, 데이터)를 작성하면 해당 이벤트가 부모컴포넌트로 전달(이벤트버블링)


- 부모컴포넌트에서는 자식컴포넌트태그에 이벤트를 받아오고, 해당 이벤트의 함수를 작성하여, 연도 1씩 증가시키는 함수 생성
- 데이터를 전달받을때는, 해당 function의 파라미터에 (event)를 추가해주면 해당 데이터가 불러와짐



- defineEmits함수를 이용해서, 발생시킨 이벤트함수를 변수명으로 선언하여, 이용하면 해당 내용의 가독성도 좋아지며, 기존의 emits에 대한 에러도 사라지게 된다.


- 가독성을 위해, 아래와 같이 구조를 그림으로 그려도 됨

    - App.vue(부모컴포넌트)에서는 state를 관리하고
    - TodoFooter... 등(자식컴포넌트)에서는 props를 관리한다.
    - 각 자식컴포넌트에서의 prop의 종류는 defineProps안에 있는 props 내용을 보면 된다.
    - 자식 -> 부모컴포넌트에게 전달하는 이벤트가 있는지 확인하는 것은, 자식 컴포넌트에 defineEmits 함수가 있는지 확인하면 알 수 있다.

## Slot (슬롯)

### 1) Slot, 기본 자식 Slots
- 슬롯, Slots
    - 컴포넌트 재사용 기법 중 하나, Component composition pattern

- 하위 컴포넌트로 HTML을 렌더링 가능

- 네임드 슬롯, Named Slots

    - 슬롯을 이용하면 컴포넌트를 다양하게 표현할 수 있는 방법의 길이 열린다

## 라우팅
- Vue Router

### 1) 라우팅이란?
- 동적 웹앱에서 중요한 부분 중 하나
    - 사용자가 원하는 위치(페이지)로 이동
    - 메뉴와 관련이 있음
        - 예)website.com/about URL 요청에 정보 페이지로 라우팅 됨
    - SPA 라우팅은 거의 동일 (Angular, React, Vue ...)
    - HTTP 요청을 처리하는 코드에 연결하는 방법을 결정하는 매칭 매커니즘
        - URL을 기반으로 원하는 리소스(페이지)에 연결해주는 기능

### 2) 라우팅: Vue Router
- Vue Router - Vue의 공식적인 라우팅 라이브러리
    - 다양한 기능으로 페이지를 새로 고침 필요없이 전환 (npm init vue)

- Vue Router 설정
    - Vue Router는 기본적으로 제공되지 않음. create-vue 설정시 옵션 선택
