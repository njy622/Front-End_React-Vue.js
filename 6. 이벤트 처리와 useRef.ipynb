# 6차시. 이벤트 처리와 useRef

### 1) DOM onclik => JSX onClick
- JSX 에서 이벤트 처리는 함수를 사용 (이벤트 핸들러는 함수 사용)<br>
    

- HTML에서 쓰는 이벤트 => Dom 이벤트 (모두 소문자로) onclick
- JSX 이벤트 => onClick
- 이벤트 핸들러는 기본적으로 함수를 사용 
    - 컴포넌트 내부에 사용할 수 있음

### 2) 이벤트 핸들러를 props로 전달
- props로 이벤트 핸들러(함수)를 전달


↑ 리액트에서 많이 사용하는 기법, 이벤트 함수 자체를 props를 넘겨주는 형태

### 3) 이벤트 전파 - bubbling, propagation
- 이벤트는 발생되는 요소에서 위로 전파 됨


↑ 버튼태그에서 이벤트가 핸들링 된 후 div 태그에서 이벤트가 2번 핸들링디는 구조

### 4) 이벤트 전파 방지
- 이벤트 핸들러는 이벤트 객체를 유일한 파라미터로 받음
    - e.stopPropagation()을 사용하여 이벤트 전파 방지 가능
- 이벤트 전파 중지와 관계없이 이벤트 캡처 가능
    - OnClickCapture 함수를 이용하면 이 이벤트의 우선순위를 가짐


### 5) DOM 직접 조작 - useRef
- 상태 변경 트리거에 의해 DOM이 변경되는 것이 React의 기본기능
    - 그러나 DOM 을 직접 조작해야 하는 경우가 생김
    - useRef 를 사용하여 DOM에 직접 접근 및 조작
    

useRef => 이벤트처리와는 별도의 기능
- 기본 적으로 컴포넌트 아키텍쳐는 DOM을 직접 변경하는것이 금지되어있음. 자동으로 변경되도록 되어있음.
- Formelemnt에서 주로 직접 변경해야하는 경우가 생김 이때, useRef사용
- Ref라는 값을 이용해서 특정 DOM에 직접 접근할 수 있으며,
해당 기법은 특수한 경우임

- onClick 이벤트 함수에 함수값을 넣어 이벤트를 처리하는 방법


- 직접 onClick함수에 직접적으로 함수의 내용을 넣어서 이용할 수도 있음


- 부모 컴포넌트에서 자식 컴포넌트로 이벤트 함수 받을때 사용법
    - 이벤트 핸들러(buttonClicked)를 다른 컴포넌트에 전달할수있다
